# 實作 server 的行為？！

input: 網址<br>
output: 把 args 印出來

```python
from urllib.parse import urlparse
url = input()
o = urlparse(url)
o.query
```

如果我想要根據不同路徑，來決定要回傳什麼呢？

```python
url = input()
o = urlparse(url)
if o.path == '/hello':
  print(o.query)
elif o.path == '/world':
  print('world')
else:
  print('index')
```
